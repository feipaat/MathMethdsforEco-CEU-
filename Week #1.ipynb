### FOLDE with constant coefficient and constant term

```octave
a=0.9;
c=0;
t=0:1:5;
x=zeros(1,size(t,2));
x(1)=-2;
for j=1:size(t,2)-1
    x(j+1)=-a*x(j)+c;
end
plot(t,x,'o--')
%title('FOLDE with const coeff and const term')
%xlabel('t time level')
%ylabel('x_t solution at time t')
```

#### Example: the growth model (national income, invest, savings)

```octave

t=0:1:5;
x=zeros(1,size(t,2));
x(1)=1;
alpha=0.2;
beta=0.8;
for j=1:size(t,2)-1
    x(j+1)=(beta/(beta-alpha)).^j*x(1);
end
plot(t,x,'o--')
title('The growth model')
xlabel('t time level')
ylabel('x_t solution at time t')
```

### SOLDE with constant coefficient and constant term

```octave
% Parameters
a1=1;
a2=-2;
c=6;
% Initial conditions
x=zeros(1,size(t,2));
x(1)=4; x(2)=5;
t=0:1:5;
% SOLDE
for j=1:size(t,2)-2
    x(j+2)=-a1*x(j+1)-a2*x(j)+c;
end
%Figure
plot(t,x,'o--')
```

```octave
% Example x_{t+2}+x_{t+1}-2x_{t}=6; x_0=4, x_1=5
% Definite Solution: x_t=(11/3)*1^t+(1/3)*(-2)^t+2*t

a1=1; a2=-2; c=6; 
t=0:1:11;
x=zeros(1,size(t,2)); x(1)=4; x(2)=5;
% General code
for j=1:size(t,2)-2
    x(j+2)=-a1*x(j+1)-a2*x(j)+c;
end
exact=(11/3)*1.^t+(1/3)*(-2).^t+2*t;
plot(t,x,'mo',t,exact);
```
